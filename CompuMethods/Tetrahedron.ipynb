# Triangle

### Area of a triangle

The area of a triangle defined by two vectors $\vec{a}$ and $\vec{b}$ is given by

$A = \frac{ab\sin{\theta}}{2} = \left |{\frac{\vec{a}\times\vec{b}}{2}}\right |$


### A point in a triangle

Given a point $\vec{d}$, one can know if it lies inside a triangle defined by vectors $\vec{a},\ \vec{b},\ \vec{c} = \vec{b} - \vec{a}$, looking the relation between the chirality in each pair of vectors. Here we take $\vec{a}$ and $\vec{b}$ as the sides of the triangle whose starting point is the origin.

Then the point $\vec{d}$ lies inside the triangle if and only if the following is satisfied:

* $(\vec{a}\times\vec{d})\cdot (\vec{d}\times\vec{c}) \geq 0$
* $(\vec{d'}\times\vec{a})\cdot (\vec{d'}\times\vec{c}) \geq 0$

where $\vec{d'}$ is the vector that points to the given point, for the common vertex $(a,c)$: 
$ \vec{d'} = \vec{d} - \vec{a}$

##### Computational proof

```python
# Determina un triangulo al azar
dim = 2
a = 1-2*np.random.random(dim)
b = 1-2*np.random.random(dim)
c = b-a
print a
```

```python
n_points = 10000
# Guarda los puntos que pasan la prueba y los que no (en arrays diferentes)
pointsIn = []
pointsOut = []
i = 0
for i in range(n_points):
    d = 1-2*np.random.random(dim)
    dp = d-a
    # Variables que guardan el signo de las quiralidades 
    quir1 = np.dot(np.cross(a,d),np.cross(d,b))
    quir2 = np.dot(np.cross(a,dp),np.cross(c,dp)) 
    if (quir1 >= 0) and (quir2 >= 0):
        pointsIn.append(d) 
    else:
        pointsOut.append(d)

pointsIn = np.array(pointsIn)
pointsOut = np.array(pointsOut)
```

### Volume of a tetrahedron

The volume of a tetrahedron defined by 3 vectors $\vec{a}$, $\vec{b}$, $\vec{c}$, is half of the volume of a parallelepiped, and is given by:

$V = \frac{abc\sin{\theta_{ab}}\sin{\theta_{ac}}}{2} = \left |\frac{(\vec{a}\times\vec{b})\cdot\vec{c}}{2}\right |$ 

### A point inside a tetrahedron 

In a similar way as in the case of a triangle, a given point $\vec{d}$ can be determined to be inside or outside a tetrahedron determined by the vectors $\vec{a}$, $\vec{b}$, $\vec{c}$, analysing the perpendicular component of that point regarding every plane formed by every pair of vectors of the tetrahedron.

Taking that into account, a point $\vec{d}$ lies inside the tetrahedron if and only if the following cuantities are the same:

1) 

* $sign(\vec{d}\cdot(\vec{a}\times\vec{b}))$
* $sign(\vec{d}\cdot(\vec{b}\times\vec{c}))$
* $sign(\vec{d}\cdot(\vec{c}\times\vec{a}))$
* $sign(\vec{d'}\cdot(\vec{e}\times\vec{f}))$


Here we defined:
+ $\vec{d'} = \vec{d} - \vec{b}$
+ $\vec{e} = \vec{a} - \vec{b}$
+ $\vec{f} = \vec{c} - \vec{b}$

```python
# Retorna el volumen de un tetraedro determinado por los vectores a,b dados
def tetraArea1(a,b,c):
    return np.linalg.norm(np.dot(np.cross(a,b),c))/2

# Retorna verdadero si el punto dado d pertenece o no al tetraedro a,b dado (retorna falso de lo contrario)
def isIn1(a,b,c,d):
    # Otros lados del tetraedro y la referencia al punto d
    e = b - a
    f = c - a
    dp = d-a
    # Variables que guardan el signo de las componentes 
    l1 = np.sign(np.dot(d,np.cross(a,b)))
    l2 = np.sign(np.dot(d,np.cross(b,c)))
    l3 = np.sign(np.dot(d,np.cross(c,a)))
    l4 = np.sign(np.dot(dp,np.cross(e,f)))
    if (l3-l4 == 0 ) and (l1-l2 == 0) and (l2-l3 == 0):
        return True
    else:
        return False
   
```

#Tetrahedron Methods II

### Volume of a tetrahedron

The volume of a tetrahedron defined by 3 vectors $\vec{a}$, $\vec{b}$, $\vec{c}$, is half of the volume of a parallelepiped, and is given by:

$V = \frac{abc\sin{\theta_{ab}}\sin{\theta_{ac}}}{2} = \left |\frac{(\vec{a}\times\vec{b})\cdot\vec{c}}{2}\right |$ 

### A point inside a tetrahedron 

Another way to validate if a point lies inside or outside a tetrahedron is a quick change of coordinates. That change of coordinates maps a vertex of the tetrahedron to (0,0,0), and the other three vertices to {(1,0,0),(0,1,0),(0,0,1)}. This can be achieved with a translation and a change of basis.

Assume that one point of the tetrahedron is alredy in (0,0,0), then we only need a change of basis. Let $\vec{a}$, $\vec{b}$ and $\vec{c}$ be the last 3 points of the tetrahedron. Then the transformation $\beta_B^{C}$ from the barycentrical coordinates to the canonical basis is given by:

$\beta_B^{C}= \{\vec{a},\vec{b},\vec{c}\} = \begin{bmatrix}{a_1}&{b_1}&{c_1}\\{a_2}&{b_2}&{c_2}\\{a_3}&{b_3}&{c_3}\end{bmatrix}$

Where:

$\vec{a} = \begin{bmatrix}{a_1}\\{a_2}\\{a_3}\end{bmatrix}$


$\vec{b} = \begin{bmatrix}{b_1}\\{b_2}\\{b_3}\end{bmatrix}$


$\vec{c} = \begin{bmatrix}{c_1}\\{c_2}\\{c_3}\end{bmatrix}$


Then, a point $\vec{d_C}$ in canonical coordinates can be expresed in the barycentrical coodrinates:

$\vec{d_B} = {\beta_B^{C}}^{-1}\vec{d_C} = \beta_C^{B}\vec{d_C} $

---
Then it is clear that a point lies inside a tetrahedron if and only if the following is satisfied:

${d_B}_1,{d_B}_2,{d_B}_3 \geq 0$

${d_B}_1 + {d_B}_2 + {d_B}_3 \leq 1$

Where:

$\vec{d_B} = \begin{bmatrix}{{d_B}_1}\\{{d_B}_2}\\{{d_B}_3}\end{bmatrix}$

```python
# Retorna el volumen de un tetraedro determinado por los vectores a,b dados
def tetraArea2(a,b,c):
    return np.linalg.norm(np.dot(np.cross(a,b),c))/2

# Retorna verdadero si el punto dado d pertenece o no al tetraedro a,b dado (retorna falso de lo contrario)
# m --> La matriz de cambio de base de canonicas a baricentricas
def isIn2(a,b,c,m,d):
    # El vector en la nueva base
    dp = np.dot(m,d)
    if (dp[0] >= 0) and (dp[1] >= 0) and (dp[2] >= 0) and (dp[0]+dp[1]+dp[2] <= 1):
        return True
    else:
        return False
```

```python
# Determina un tetraedro al azar
dim = 3
tam = 1
a = tam*(1-2*np.random.random(dim))
b = tam*(1-2*np.random.random(dim))
c = tam*(1-2*np.random.random(dim))
print a
```

```python
n_points = 100000
# Guarda los puntos que pasan la prueba1 y los que no (en arrays diferentes)
pointsIn1 = []
pointsOut1 = []
# Guarda los puntos que pasan la prueba2 y los que no (en arrays diferentes)
pointsIn2 = []
pointsOut2 = []
# Metodo 1
start_time = time.time()
for i in range(n_points):
    d = tam*(1-2*np.random.random(dim))
    if isIn1(a,b,c,d):
        pointsIn1.append(d) 
    else:
        pointsOut1.append(d)
print(time.time() - start_time)
# Metodo 2
start_time = time.time()
m = np.linalg.inv(np.array([a,b,c]).T)
for i in range(n_points):
    d = tam*(1-2*np.random.random(dim))
    if isIn2(a,b,c,m,d):
        pointsIn2.append(d) 
    else:
        pointsOut2.append(d)
print(time.time() - start_time)        

pointsIn1 = np.array(pointsIn2)
pointsOut1 = np.array(pointsOut2)
pointsIn2 = np.array(pointsIn2)
pointsOut2 = np.array(pointsOut2)
```
