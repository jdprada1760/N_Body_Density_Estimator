# Triangle

### Area of a triangle

The area of a triangle defined by two vectors $\vec{a}$ and $\vec{b}$ is given by

$A = \frac{ab\sin{\theta}}{2} = \left |{\frac{\vec{a}\times\vec{b}}{2}}\right |$


### A point in a triangle

Given a point $\vec{d}$, one can know if it lies inside a triangle defined by vectors $\vec{a},\ \vec{b},\ \vec{c} = \vec{b} - \vec{a}$, looking the relation between the chirality in each pair of vectors. Here we take $\vec{a}$ and $\vec{b}$ as the sides of the triangle whose starting point is the origin.

The point $\vec{d}$ lies then inside the triangle if and only if the following is satisfied:
* $(\vec{a}\times\vec{d})\cdot (\vec{d}\times\vec{c}) \geq 0$
* $(\vec{d'}\times\vec{a})\cdot (\vec{d'}\times\vec{c}) \geq 0$

where $\vec{d'}$ is the vector that points to the given point, for the common vertex $(a,c)$: 
$ \vec{d'} = \vec{d} - \vec{a}$

##### Computational proof

```python
# Determina un triangulo al azar
dim = 2
a = 1-2*np.random.random(dim)
b = 1-2*np.random.random(dim)
c = b-a
print a
```

```python
n_points = 10000
# Crea un array de vectores para determinar si caen dentro del triangulo abc
points = 1-2*np.random.random((n_points,dim))
# Guarda los puntos que pasan la prueba y los que no (en arrays diferentes)
pointsIn = []
pointsOut = []
i = 0
for i in range(n_points):
    d = points[i]
    dp = d-a
    # Variables que guardan el signo de las quiralidades 
    quir1 = np.dot(np.cross(a,d),np.cross(d,b))
    quir2 = np.dot(np.cross(a,dp),np.cross(c,dp)) 
    if (quir1 >= 0) and (quir2 >= 0):
        pointsIn.append(d) 
    else:
        pointsOut.append(d)

pointsIn = np.array(pointsIn)
pointsOut = np.array(pointsOut)
```
