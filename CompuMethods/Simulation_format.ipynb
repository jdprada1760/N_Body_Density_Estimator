#Relation
From above it is clear that for the first $16*16*16$ particles there is a relation between their coordinates $(i,j,k)$ and $N$, their Id.

### $N \rightarrow (i,j,k)$
We define the length of the side of the cube as $L$:

$L = 16$

Then the coordinates are given by:

$ k = N\%L $

$ j = P(N/L)\%L $

$ i = P(N/L^2)  $

Where $P(x)$ is the floor integer component function

### $(i,j,k) \rightarrow N $
Clearly:

$ N = iL^2 + jL + k $

--------------------

## Cube defined by a point $(i,j,k)$

We define a cube as the set of points:

$P_1 = (i,j,k) \rightarrow N$

$P_2 = (i+1,j,k)\rightarrow N + L^2$

$P_3 = (i,j,k+1)\rightarrow N +1 $

$P_4 = (i,j+1,k)\rightarrow N + L$

$P_5 = (i+1,j+1,k)\rightarrow N + L^2 +L$

$P_6 = (i+1,j,k+1)\rightarrow N + L^2 + 1$

$P_7 = (i,j+1,k+1)\rightarrow N + L + 1$

$P_8 = (i+1,j+1,k+1)\rightarrow N + L^2 + L + 1$

----------------

### Tetrahedrons in the cube $(i,j,k)$

We define now 6 tetrahedrons, 4 set of points each (4 sets of Ids), in that given cube:

$T_1 =  \{ P_1, P_4, P_5, P_8 \} \rightarrow N + \{ 0, L, L^2 +L, L^2 +L +1\} $

$T_2 =  \{ P_1, P_4, P_7, P_8 \} \rightarrow N + \{ 0, L, L +1, L^2 +L +1\} $

$T_3 =  \{ P_1, P_2, P_5, P_8 \} \rightarrow N + \{ 0, L^2, L^2 +L, L^2 +L +1\} $

$T_4 =  \{ P_1, P_2, P_6, P_8 \} \rightarrow N + \{ 0, L^2, L^2 +1, L^2 +L +1\} $

$T_5 =  \{ P_1, P_3, P_7, P_8 \} \rightarrow N + \{ 0, 1, L +1, L^2 +L +1\} $

$T_6 =  \{ P_1, P_3, P_6, P_8 \} \rightarrow N + \{ 0, 1, L^2 +1, L^2 +L +1\} $

```python
# Obtiene un punto al azar y define los puntos del cubo asociado a el
p1 = np.random.random(3)
p2 = p1 + np.array([1,0,0])
p3 = p1 + np.array([0,0,1])
p4 = p1 + np.array([0,1,0])
p5 = p1 + np.array([1,1,0])
p6 = p1 + np.array([1,0,1])
p7 = p1 + np.array([0,1,1])
p8 = p1 + np.array([1,1,1])
cube = np.array([p1,p2,p3,p4,p5,p6,p7,p8])
```

```python
# Define los tetraedros asociados a ese cubo
T1 = np.array([p1,p4,p5,p8])
T2 = np.array([p1,p4,p7,p8])
T3 = np.array([p1,p2,p5,p8])
T4 = np.array([p1,p2,p6,p8])
T5 = np.array([p1,p3,p7,p8])
T6 = np.array([p1,p3,p6,p8])
```

```python
# Grafica el tetraedro que entra por parámetro con el color dado
def TetraGraph(Th, color):
     # Numero de particulas a mostrar n1,n2
    from mpl_toolkits.mplot3d import Axes3D
    fig = pylab.figure()
    ax = Axes3D(fig)
    ax.plot(Th.T[0][[0,1]],Th.T[1][[0,1]],Th.T[2][[0,1]], c = color, linewidth = 2.5)
    ax.plot(Th.T[0][[0,2]],Th.T[1][[0,2]],Th.T[2][[0,2]], c = color, linewidth = 2.5)
    ax.plot(Th.T[0][[0,3]],Th.T[1][[0,3]],Th.T[2][[0,3]], c = color, linewidth = 2.5)
    ax.plot(Th.T[0][[1,2]],Th.T[1][[1,2]],Th.T[2][[1,2]], c = color, linewidth = 2.5)
    ax.plot(Th.T[0][[1,3]],Th.T[1][[1,3]],Th.T[2][[1,3]], c = color, linewidth = 2.5)
    ax.plot(Th.T[0][[2,3]],Th.T[1][[2,3]],Th.T[2][[2,3]], c = color, linewidth = 2.5)
    ax.scatter(Th.T[0],Th.T[1],Th.T[2], c = 'b', s = 10)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
```

```python
# Grafica todos los tetraedros y el cubo
%pylab inline
TetraGraph(T1, 'r')
TetraGraph(T2, 'g')
TetraGraph(T3, 'b')
TetraGraph(T4, 'r')
TetraGraph(T5, 'g')
TetraGraph(T6, 'b')
```
