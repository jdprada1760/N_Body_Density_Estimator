```python
# Grafica 3D
%pylab inline
def graf(n1,n2):
    # Numero de particulas a mostrar n1,n2
    from mpl_toolkits.mplot3d import Axes3D
    fig = pylab.figure()
    ax = Axes3D(fig)
    #ax.scatter(initial.T[0][0:n1],initial.T[1][0:n1],initial.T[2][0:n1], c = 'b')
    ax.scatter(initial.T[0][n1:n2],initial.T[1][n1:n2],initial.T[2][n1:n2], c = 'r')
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
```
